```python
from datetime import datetime
def str_to_datetime(x, sep='.'):
    try:
        result = datetime(*(map(lambda x:int(x),x.split(sep))))
    except:
        result = datetime(*(map(lambda x:int(x),x.split('-'))))
    return result
```

```python
receipt_map = {
           '1ë…„ ì˜ìˆ˜ì¦ ê°œìˆ˜' : [0, '2019.06.23'],
           '6ê°œì›” ì˜ìˆ˜ì¦ ê°œìˆ˜' : [0, '2019.12.23'],
           '3ê°œì›” ì˜ìˆ˜ì¦ ê°œìˆ˜' : [0, '2020.03.23'],
           '2ê°œì›” ì˜ìˆ˜ì¦ ê°œìˆ˜' : [0, '2020.04.23'],
           '1ê°œì›” ì˜ìˆ˜ì¦ ê°œìˆ˜' : [0, '2020.05.23']}
receipt_scores = ['1ë…„ ì ìˆ˜', '6ê°œì›” ì ìˆ˜', '3ê°œì›” ì ìˆ˜', '2ê°œì›” ì ìˆ˜', '1ê°œì›” ì ìˆ˜']
def custom(each_df):
    
#     print(each_df.iloc[:,0])
    new_df = {x:0 for x in receipt_map.keys()}
    [new_df.update({x: 0}) for x in receipt_scores]
#     new_df['loc_name'] = each_df.iloc[0, 0]
    for date, score in zip(each_df['dates'].values, each_df['scores'].values):
        for idx, (k, values) in enumerate(receipt_map.items()):
            if str_to_datetime(date) > str_to_datetime(values[1]):
                new_df[k] += 1
                new_df[np.array(receipt_scores)[list(map(lambda x:k.split(' ')[0] in x, receipt_scores))][0]] += float(score)
    for score_col, num_col in zip(receipt_scores, receipt_map.keys()):
        new_df[score_col] = new_df[score_col] / new_df[num_col]
    return pd.Series(new_df)
```
