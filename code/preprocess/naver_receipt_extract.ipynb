```python

for idx, row in naver_reveiw.iterrows():
    receipt_map = {
               '1년 영수증 개수' : [0, '2019.06.23'],
               '6개월 영수증 개수' : [0, '2019.12.23'],
               '3개월 영수증 개수' : [0, '2020.03.23'],
               '2개월 영수증 개수' : [0, '2020.04.23'],
               '1개월 영수증 개수' : [0, '2020.05.23']}
    receipt_score = ['1년 점수', '6개월 점수', '3개월 점수', '2개월 점수', '1개월 점수']
    scores = [0, 0, 0, 0, 0]
    row_scores = ast.literal_eval(row['scores'])
#     print(f'receipt number after {base_date} : {cnt}')
    for k, date in ast.literal_eval(row['dates']).items():
        for i, (map_col, date_info) in enumerate(receipt_map.items()):
            if str_to_datetime(date) >= str_to_datetime(date_info[1]):
                date_info[0]+=1
                scores[i]+=float(row_scores[k])
    for i, (map_col, date_info) in enumerate(receipt_map.items()):
        naver_reveiw.loc[idx, map_col] = date_info[0]
        naver_reveiw.loc[idx, receipt_score[i]] = scores[i] / date_info[0]
    
#     print(f'영수증 개수 : {naver_reveiw.loc[idx, list(receipt_map.keys())]}')
```
