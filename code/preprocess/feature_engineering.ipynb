```python
def agg_mean(x):
    total = 0
    cnt = 0
    for year in [2017, 2018, 2019]:
        if not np.isnan(x[year]):
            total += x[year]
            cnt += 1
    return total / cnt

def agg_moving_average(x):
    diff = 0
    cnt = 0
    if np.array([np.isnan(x[y]) for y in [2017, 2018, 2019]]).any():
        return 0
    for year in [2019, 2018]:
        if np.isnan(x[year]):
            break
        elif not np.isnan(x[year-1]):
            diff += x[year] - x[year-1]
            cnt += 1
    return diff / cnt

def agg_diff(x):
    diff = 0
    cnt = 0
    if np.array([np.isnan(x[y]) for y in [2017, 2018, 2019]]).any():
        return 0
    for year in [2019, 2018]:
        if np.isnan(x[year]):
            break
        elif not np.isnan(x[year-1]):
            diff += x[year] - x[year-1]
            cnt += 1
    return diff

new_price_3year_gp_pivot['3year'] = new_price_3year_gp_pivot.apply(lambda x:agg_mean(x), axis=1)
new_price_3year_gp_pivot['3year_diff_mean'] = new_price_3year_gp_pivot.apply(lambda x:agg_moving_average(x), axis=1)
new_price_3year_gp_pivot['3year_diff'] = new_price_3year_gp_pivot.apply(lambda x:agg_diff(x), axis=1)
```

```python
def column_list_to_multiple_rows(df, base_columns: list, multi_column_names: list):
    df_dict = {}
    for base_column in base_columns:
        df_dict[base_column] = np.repeat(df[base_column].values, df[multi_column_names[0]].str.len())
    for idx, column_name in enumerate(multi_column_names):
        df_dict[column_name] = np.concatenate(df[column_name].values)
    return pd.DataFrame(df_dict)
```

```python
import pandas as pd
import geopy.distance


def get_distances_with_df(df1, df2, df1_basis, df2_basis):
    rows = []
    for j in range(df1.shape[0]):
        loc_name1 = df1.loc[j][df1_basis]
        coord1 = (df1.loc[j].latitude, df1.loc[j].longitude)
        for i in range(df2.shape[0]):
            loc_name2 = df2.loc[i][df2_basis]
            coord2 = (df2.loc[i].latitude, df2.loc[i].longitude)
            distance = geopy.distance.vincenty(coord1, coord2).km
            row = {
                df1_basis : loc_name1,
                df2_basis : loc_name2,
                'distance': distance
            }
            rows.append(row)
    return rows


def find_mindist_starbucks(groupped_df, basis_name='star_name'):
    new_df = {}
#     groupped_df = groupped_df.reset_index(drop=True)
    groupped_df = groupped_df.sort_values(by='distance').reset_index(drop=True)
    new_df['distance'] = groupped_df.loc[:2]['distance']
    new_df[basis_name] = groupped_df.loc[:2][basis_name]

    return pd.Series(new_df, index=[basis_name, 'distance'])



def main():
    loc_info = pd.read_csv('../data/mac/ST_MAC2_DT.csv', sep='\t')
    loc_info_mac = loc_info[loc_info.SORT == '맥도날드']
    loc_info_star = loc_info[loc_info.SORT == '스타벅스']
    loc_info_star = loc_info_star.reset_index().drop('index', axis=1)

    loc_info_mac = loc_info_mac.drop_duplicates().reset_index(drop=True)

    rows = get_distances_with_st(loc_info_mac, loc_info_star)
    mac_star_distance = pd.DataFrame(rows)
    mac_star_distance2 = mac_star_distance.groupby(['mac_name']).apply(find_mindist_starbucks)
    mac_star_distance3 = mac_star_distance2.reset_index()
    mac_star_distance3.to_csv('data/mac_star_distance2.csv', sep='\t')

# if __name__ == '__main__':;;
#     main()
```
